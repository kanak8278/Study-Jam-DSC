```python
import pandas as pd
```

```python
ipl_df = pd.read_csv(f"{location}/IPL2020.csv")
```

```python
different_matches = ipl_df['match_id'].unique()
```

```python
ipl_df.head()
```

```python
match_id = []
innings1 = []
innings2 = []
first_inning_team= []
second_inning_team= []
for x in different_matches:
    inning1 = ipl_df[(ipl_df['match_id'] == x) & (ipl_df["inning"]==1)]['total_runs'].sum()
    inning2 = ipl_df[(ipl_df['match_id'] == x) & (ipl_df["inning"]==2)]['total_runs'].sum()
    team1 = ipl_df[(ipl_df['match_id'] == x) & (ipl_df["inning"]==1)]['batting_team'].unique()[0]
    team2 = ipl_df[(ipl_df['match_id'] == x) & (ipl_df["inning"]==2)]['batting_team'].unique()[0]
    match_id.append(x)
    innings1.append(inning1)
    innings2.append(inning2)
    first_inning_team.append(team1)
    second_inning_team.append(team2)
    
    
    
```

```python
len(innings1)
```

```python
data_dict ={
    "match_id":match_id,
    "innings1_team": first_inning_team,
    "innings2_team": second_inning_team,
    "innings1_score": innings1,
    "innings2_score": innings2,
}
```

```python
resultant_Data = pd.DataFrame.from_dict(data_dict)
```

```python
resultant_Data
```

```python
resultant_Data['binary'] = resultant_Data['innings1_score']>resultant_Data['innings2_score']
```

```python
winner = []
for i, val in enumerate(resultant_Data['binary']):
    if(val== False):
        winner.append(resultant_Data.loc[i]['innings2_team'])
    else:
        winner.append(resultant_Data.loc[i]['innings1_team'])
```

```python
len(winner)
```

```python
resultant_Data['winner'] = winner
```

```python
resultant_Data.iloc[35]
```

```python
final_Result = resultant_Data.drop("binary", axis=1)
```

```python
final_Result.to_csv("IPL_Win_Loss.csv")
```
